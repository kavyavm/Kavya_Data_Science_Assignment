### Visualizations using the dataset

### 1).Correlation

```python
# Finding the correlation between each columns in the dataset using corr method

df.corr()
```

### Inference
#### 1. Petal length and Petal width have high correlations
#### 2. Sepal length and Petal width have good correlations
#### 3. Petal length and Sepal length have good correlations

### 2).Countplot for Classification

```python
sns.countplot(x = 'Classification',data = df )
plt.grid()
```

### Inference
##### Countplot shows the count of all three species Iris-Setosa,Iris-versicolor and Iris-virginica are same(count = 50)

###  3).Scatterplot
### 3.1.Comparing SL and SW using Scatterplot

```python
sns.scatterplot(x = 'SL', y = 'SW', hue = 'Classification',data = df)
# Placing legend outside the figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.title("Plot of SL Vs SW")
plt.grid()
plt.show()
```

### Inference
 #### 1. Species Setosa has smaller sepal lengths and larger sepal widths
 #### 2. Species Versicolor lies in the middle of other two species in terms of sepal length and sepal width
 #### 3. Species Virginica has larger sepal lengths and smaller sepal widths

### 3.2.Comparing PL and PW using Scatterplot

```python
sns.scatterplot(x = 'PL', y = 'PW', hue = 'Classification',data = df)
# Placing legend outside the figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.title("Plot of PL Vs PW")
plt.grid()
plt.show()
```

### Inference
 #### 1. Species Setosa has smaller petal length and petal width
 #### 2. Species Versicolor lies in the middle of other two species in terms of petal length and petal width
 #### 3. Species Virginica has largest petal length and petal width

### 4).Bar Chart

```python
plt.bar(df['Classification'],df['PL'],color = 'maroon')
plt.xlabel('Classification')
plt.ylabel('PL')
plt.title('Plot of Classification Vs PL',fontsize = 14)
plt.show()
```

### Inference
#### 1.Bar chart shows that PL of species Iris-virginica is greatest when compared to other two.
#### 2.Petal length of species Iris- setosa is lowest when compared to others.

### 5).Density Plot

```python
df.plot.kde()
plt.show()
```

### Inference
#### 1. From the above graph it is understood that SW has higher density compared to SL,PL and PW.
#### 2. Petal length has lowest intensity when compared to other features.

### 6). Pair plot

```python
sns.pairplot(df,hue = 'Classification')
plt.show()
```

### Inference
#### 1. Values of Iris-Setosa are distinct from Iris-versicolor and Iris-Virginica.So it is easy to distinguish iris-setosa 
#### from other two species.
#### 2. Iris-Versicolor and Iris-Virginica have overlapping values
