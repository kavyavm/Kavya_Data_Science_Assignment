### 1). The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.

### Step 1 : Defining null hypothesis(H0) and alternate hypothesis(HA)

#### H0 - There is no change in sales after stepping into digital marketing
#### HA - There is increase in sales after stepping into digital marketing

### Step 2 : Deciding which test to use 
#### Here sample size is less than 30; So T-test is used to calculate p value (A T- test is a statistical test that is used to compare the means of two groups)

```python
# Importing necessary libraries for performing two-tailed t-test
import scipy.stats as stats
from scipy.stats import ttest_ind
```

### Step 3 : Calculating p value to decide whether null hypothesis is rejected or accepted

```python
t_test,p_value = ttest_ind(data['Sales_before_digital_add(in $)'],data['Sales_After_digital_add(in $)'])
```

```python
p_value
```

```python
# Level of significance (alpha) = 0.05
# Comparing p value with alpha to decide which hypothesis is accepted
if p_value < 0.05:
    print('Null hypothesis is rejected')
else:
    print('Null hypothesis is accepted')
```

### Step 4 : Conclusion
#### Since p value is less than alpha value,null hypothesis is rejected.Hence we can say that there is increase in sales after stepping into digital marketing

### 2).The company needs to check whether there is any dependency between the features 'Region' and 'Manager'

### Step 1 : Defining null hypothesis(H0) and alternate hypothesis(HA)

#### H0 - There is no dependency between the features 'Region' and 'Manager'
#### HA - There is dependency between the features 'Region' and 'Manager'

### Step 2 : Deciding which test to use 
#### Here since we have to find the association between two categorical variables,chi-square test of independence can be used. (The chi-square test of independence also known as the chi-square test of association which is used to determine the association between the categorical variables).

```python
# Importing necessary libraries to perform chi-square test of independence
from scipy.stats import chi2_contingency
```

### Step 3 : Calculating p value to decide whether null hypothesis is rejected or accepted

```python
data_tabulation = pd.crosstab(data['Region'],data['Manager'])
data_tabulation
```

```python
chi_stat,p_val,dof,expected = chi2_contingency(data_tabulation)
```

```python
p_val
```

```python
# Level of significance (alpha) = 0.05
# Comparing p value with alpha to decide which hypothesis is accepted
if p_val < 0.05:
    print('Null hypothesis is rejected')
else:
    print('Null hypothesis is accepted')
```

### Step 4 : Conclusion
#### Since p value is greater than alpha,null hypothesis is accepted.That means we can conclude that there is no dependency between the features 'Region' and 'Manager'
